## 3. ***Data Wrangling***

### Data Wrangling Code

 Giving more approproate column names

```python
airbnb_df = airbnb_df.rename(columns = {'neighbourhood_group':'borough', 'last_review':'date_of_last_review','reviews_per_month':'average_reviews_per_month'})
airbnb_df
```

```python
airbnb_df.isnull().sum()
```

Filling "Unknown" in place of missing host, listing names and last date of review

```python
airbnb_df["name"].fillna("Unknown", inplace = True)
airbnb_df["host_name"].fillna("Unknown", inplace = True)
airbnb_df["date_of_last_review"].fillna("Unknown", inplace = True)
airbnb_df[["name","host_name","date_of_last_review"]].isna().sum()
```

Fill in missing values in average reviews per month with the average of the remaining values

```python
av_rev = airbnb_df["average_reviews_per_month"].mean(skipna = True)
airbnb_df["average_reviews_per_month"].fillna(av_rev, inplace = True)
airbnb_df[["average_reviews_per_month"]].isna().sum()
```

### List of Manipulations

Column name had 16 null valures, host_name had 21 null values and average_reviews_per_month abd date_of_last_review had 10052 reviews per month.
1. My first manipulation was to change three column names; neighbourhood_goup to borough, last_review to date of last_review and reviews_per_month to average_reviews_per_month
2. My second manipulation involved filling null values in name, host_name and date_of_last_review with "Unknown".
3. My final manipulation involved filling null values in average_reviews per_month with the mean of the remaining values in the column
